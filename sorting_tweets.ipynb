```python
import pandas as pd
df=pd.read_csv("TweetsOne.csv")
df.head()
```

```python
def success_category(row):
    if row.retweet_count < 100:
        return "Low"
    elif row.retweet_count > 100 and row.retweet_count < 200:
        return "Medium"
    elif row.retweet_count > 200:
        return "High"
```

```python
categories = df.apply(lambda row: success_category(row), axis=1)

new_df = pd.DataFrame()
new_df["text"] = df["tweet_text"]
new_df["retweet_count"] = df["retweet_count"]
new_df["success_category"] = categories
```

```python
new_df
```

```python
new_df[new_df["success_category"] == "High"].to_csv("high_tweets.csv", index=False)
new_df[new_df["success_category"] == "Medium"].to_csv("medium_tweets.csv", index=False)
new_df[new_df["success_category"] == "Low"].to_csv("low_tweets.csv", index=False)


```
