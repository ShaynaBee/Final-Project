```python
import tweepy
from config import bearer_token

client = tweepy.Client(bearer_token)
```

```python
num_loops = 10
num_results_per_loop = 10
next_token = None

ids = []
tweet_texts = []
public_metrics = []
created_times = []
sources = []
languages = []

for i in range(num_loops):
    print(f"Request {i}    ===========")
    
    response = client.search_recent_tweets("data science", 
                                           max_results=num_results_per_loop, 
                                           next_token=next_token,
                                           tweet_fields=['id', 'text', 'author_id', 'created_at', 'public_metrics','source', 'lang'])
    # The method returns a Response object, a named tuple with data, includes,
    # errors, and meta fields
    print(response.meta)

    # In this case, the data field of the Response returned is a list of Tweet
    # objects
    tweets = response.data
    
    # Each Tweet object has default ID and text fields
    for tweet in tweets:
        ids.append(tweet.id)
        tweet_texts.append(tweet.text)
        created_times.append(tweet.created_at)
        sources.append(tweet.source)
        public_metrics.append(tweet.public_metrics)
        languages.append(tweet.lang)
        
        
        
    next_token = response.meta['next_token']
    
    print()
```

```python
import pandas as pd
df = pd.DataFrame(
    {'id': ids, 
     'tweet_text': tweet_texts, 
     'created_at': created_times,
     'source': sources,
     'public_metrics':public_metrics,
     'language': languages
    })
df
```

```python
normalized_json = pd.json_normalize(df.public_metrics)
normalized_json
```

```python
merged_df = pd.merge(df, normalized_json, left_index=True, right_index=True)
merged_df
```
